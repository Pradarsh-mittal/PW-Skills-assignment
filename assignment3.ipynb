```python
# 1. Difference between a Function and a Method in Python

# Function: A block of code that performs a task, defined using def. It is not tied to an object.

# Method: A function that is defined inside a class and is associated with an object. It takes the object instance (self) as its first parameter.
```

```python

def add(a, b):
    return a + b

print(add(2, 3))
class Calculator:
    def multiply(self, x, y):
        return x * y

obj = Calculator()
print(obj.multiply(4, 5))
```

```python
# 6. Generators

# A generator is a function that uses yield instead of return. It produces values lazily (one at a time).
```

```python
def my_gen():
    yield 1
    yield 2
    yield 3

for val in my_gen():
    print(val)
```

```python
# 7. Advantages of Generators

# Memory efficient (don’t store all values at once).

# Lazy evaluation (generate values on demand).

# Infinite sequence support.
```

```python
def infinite_numbers():
    n = 1
    while True:
        yield n
        n += 1

gen = infinite_numbers()
print(next(gen))  # 1
print(next(gen))  # 2
```

```python
# 8. Lambda Function

# An anonymous function (no name) written with lambda.

# Often used for short, one-time operations.

# Example:

square = lambda x: x * x
print(square(7))
```

```python
# 9. Purpose of map()

# Applies a function to each element of an iterable.

# Example:

nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
print(squares)
```

```python
# 10. Difference between map(), reduce(), and filter()

# map() → applies a function to each item → returns transformed list.

# filter() → filters items based on a condition.

# reduce() → applies a function cumulatively to reduce a list into a single value.

# Example:

from functools import reduce

nums = [47, 11, 42, 13]
print(list(map(lambda x: x*2, nums)))

print(list(filter(lambda x: x % 2 == 0, nums)))
print(reduce(lambda x, y: x + y, nums))
```

```python
# 11. Internal Mechanism of reduce() for Sum

# Given list: [47, 11, 42, 13]

from functools import reduce
nums = [47, 11, 42, 13]
result = reduce(lambda x, y: x + y, nums)
print(result)
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
primes = [x for x in range(1,201) if is_prime(x)]
print(primes)
```

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b, self.count = 0, 1, 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count < self.n:
            val = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return val
        else:
            raise StopIteration
for i in Fibonacci(10):
    print(i, end=" ")
print()
```

```python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i
for p in powers_of_two(5):
    print(p, end=" ")
print()

```
